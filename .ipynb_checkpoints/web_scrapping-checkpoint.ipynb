```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
```

```python
import time
time.sleep(10)
```

```python
service = Service("chromedriver.exe")
driver = webdriver.Chrome(service=service)
```

```python
url = "https://internshala.com/internships/python-internship"
driver.get(url)
time.sleep(5)
```

```python
print(driver.title)
```

```python
page_source = driver.page_source
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page_source, "html.parser")
```

```python
internships = soup.find_all(
    "div",
    class_="individual_internship"
)

real_internships = [
    i for i in internships
    if "jos_native_ad_text" not in i.get("class", [])
]

print("Real internships:", len(real_internships))

```

```python
print(real_internships[0].prettify())
```

```python
first = real_internships[0]

title = first.find("a", class_="job-title-href").get_text(strip=True)
company = first.find("p", class_="company-name").get_text(strip=True)
location = first.find("div", class_="locations").get_text(" ", strip=True)
stipend = first.find("span", class_="stipend").get_text(strip=True)

print(title)
print(company)
print(location)
print(stipend)
```

```python
data = []

for internship in real_internships:
    try:
        title = internship.find("a", class_="job-title-href").get_text(strip=True)
        company = internship.find("p", class_="company-name").get_text(strip=True)
        location = internship.find("div", class_="locations").get_text(" ", strip=True)
        stipend = internship.find("span", class_="stipend").get_text(strip=True)

        # duration (safe)
        duration_tag = internship.find("i", class_="ic-16-calendar")
        duration = duration_tag.find_next("span").get_text(strip=True) if duration_tag else None

        # apply link
        apply_link = "https://internshala.com" + internship.get("data-href", "")

        data.append({
            "Title": title,
            "Company": company,
            "Location": location,
            "Stipend": stipend,
            "Duration": duration,
            "Apply Link": apply_link
        })

    except Exception:
        continue
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.head()
```

```python
df.to_excel("internshala_internships.xlsx", index=False)
```
